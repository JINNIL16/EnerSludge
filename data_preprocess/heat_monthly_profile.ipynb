```python
import os 
import pandas as pd 
import numpy as np
```

```python
df_amount = pd.read_excel(r"E:\Test\Inputs\heat_supply_2020.xlsx")
df_m = pd.read_csv(r"E:\Test\Inputs\heat_supply_month_province.csv")
```

```python
df_m
```

```python
m_with = df_amount.set_index('Province')['monthly_supply_with'].to_dict()

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
for month in months:
    df_m[month] = df_m.apply(lambda row: row[month] * m_with[row['Province']], axis=1)
```

```python
m_without = df_amount.set_index('Province')['monthly_supply_without'].to_dict()

for month in months:
    df_m[month] = df_m.apply(lambda row: row[month] if not np.isnan(row[month]) else m_without[row['Province']], axis=1)
```

```python
df_m = df_m.drop(columns=['Total'])
```
