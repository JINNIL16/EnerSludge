## distance_province

```python
# 提取 V1 和 V2 列中的前两个数字
CFPP['V1_prefix'] = CFPP['V1']// 100000000
CFPP['V2_prefix'] = CFPP['V2']// 100000000

# 比较前两个数字是否相同
mask = ((CFPP['V1_prefix'] != 65) & (CFPP['V1_prefix'] == CFPP['V2_prefix'])) | ((CFPP['V1_prefix'] == 65) | (CFPP['V1_prefix'] == 69)) & (CFPP['V2_prefix'] == 65)

# 过滤数据，保留符合条件的数据
CFPP_filtered = CFPP[mask]

# 删除添加的前缀列
CFPP_filtered.drop(columns=['V1_prefix', 'V2_prefix'], inplace=True)

CFPP_filtered.to_csv('distance_cfpp_province.csv',index=False)
```

## 生成FID

```python
folder_path = r"E:/Test/Inputs/"
folder_path_1 = r"E:/Test/"

# WWTP
file_WWTP = os.path.join(folder_path, "wastewater_treatment_plants_5218.xlsx")
df_WWTP = pd.read_excel(file_WWTP)

# WIPP
file_WIPP = os.path.join(folder_path, "waste_inceration_plants_2020_operating.xlsx")
df_WIPP = pd.read_excel(file_WIPP)

file_WIPP_distance_1 = os.path.join(folder_path_1, "distance_wipp_nation.csv")
df_WIPP_dis_all = pd.read_csv(file_WIPP_distance_1)

file_WIPP_distance_2 = os.path.join(folder_path_1, "distance_wipp_province.csv")
df_WIPP_dis_pro = pd.read_csv(file_WIPP_distance_2)
```

```python
FID_WWTP = dict(zip(df_WWTP['项目编号'], df_WWTP['FID']))
FID_WIPP = dict(zip(df_WIPP['Plant_Code'], df_WIPP['FID']))


for index, row in df_WIPP_dis_all.iterrows():
    wwtp_code = row['V1']
    code = row['V2']
    
    if wwtp_code in FID_WWTP:
        INPUT_FID = FID_WWTP[wwtp_code]
    
    if code in FID_WIPP:
        NEAR_FID = FID_WIPP[code]
    
    df_WIPP_dis_all.at[index,'INPUT_FID'] = INPUT_FID
    df_WIPP_dis_all.at[index,'NEAR_FID'] = NEAR_FID

df_WIPP_dis_all.to_csv('distance_wipp_nation.csv',index=False)
```

```python
FID_WWTP = dict(zip(df_WWTP['项目编号'], df_WWTP['province_FID']))
FID_WIPP = dict(zip(df_WIPP['Plant_Code'], df_WIPP['province_FID']))
province = dict(zip(df_WIPP['Plant_Code'], df_WIPP['Pro_en']))


for index, row in df_WIPP_dis_pro.iterrows():
    wwtp_code = row['V1']
    code = row['V2']
    
    if wwtp_code in FID_WWTP:
        INPUT_FID = FID_WWTP[wwtp_code]
    
    if code in FID_WIPP:
        NEAR_FID = FID_WIPP[code]
        Province = province[code]
    
    df_WIPP_dis_pro.at[index,'INPUT_FID'] = INPUT_FID
    df_WIPP_dis_pro.at[index,'NEAR_FID'] = NEAR_FID
    df_WIPP_dis_pro.at[index,'Province'] = Province

df_WIPP_dis_pro.to_csv('distance_wipp_province.csv',index=False)
```

## 筛选距离

```python
# 定义距离上限
distance_threshold = 450
```

```python
# 文件路径
# file_path = r"E:\Test\Inputs\distance_cfpp_wihout_ot_nation.csv"
file_path = r"E:\Test\Inputs\distance_cfpp_nation.csv"
data_cfpp = pd.read_csv(file_path)

# 筛选数据
cfpp_filtered = data_cfpp[data_cfpp['distance'] <= distance_threshold]

# 保存到CSV文件
# output_file = f'E:/Test/Inputs/distance_cfpp_nation_P2_{distance_threshold}.csv'
output_file = f'E:/Test/Inputs/distance_cfpp_nation_{distance_threshold}.csv'
cfpp_filtered.to_csv(output_file, index=False)
```
