```python
import pandas as pd
import itertools
```

```python
df = pd.read_csv(r"E:\Test\Inputs\costs\df_NC_summary_100.csv")

if 'Unnamed: 0' in df.columns:  # 假设第一列是行标签（例如 'mean', 'std', 'min', 等）
    df.set_index('Unnamed: 0', inplace=True)
```

```python
df
```

```python
# 提取的值
first_five_min = df.loc['Lower Bound', ['NC_landappli', 'NC_building', 'NC_inci', 'NC_landfill', 'NC_other']]
first_five_ave = df.loc['mean', ['NC_landappli', 'NC_building', 'NC_inci', 'NC_landfill', 'NC_other']]
first_five_max = df.loc['Upper Bound', ['NC_landappli', 'NC_building', 'NC_inci', 'NC_landfill', 'NC_other']]

coal_min = df.loc['Lower Bound', 'TC_coal']
coal_ave = df.loc['mean', 'TC_coal']
coal_max = df.loc['Upper Bound', 'TC_coal']

msw_min = df.loc['min', 'NC_cowaste']
msw_ave = df.loc['mean', 'NC_cowaste']
msw_max = df.loc['Upper Bound', 'NC_cowaste']

# 组合前五种方案的 MIN、AVE、MAX
first_five_combinations = [
    first_five_min.values,
    first_five_ave.values,
    first_five_max.values
]

# 组合后两种方案的 MIN、AVE、MAX
last_two_combinations = list(itertools.product(
    [coal_min, coal_ave, coal_max],
    [msw_min, msw_ave, msw_max]
))

# 生成所有组合
combinations = list(itertools.product(first_five_combinations, last_two_combinations))

# 展平组合并创建 DataFrame
flattened_combinations = [
     list(five) + list(two) for five, two in combinations
]

# 创建 DataFrame
columns = [
    "NC_landappli", "NC_building", "NC_inci", 
    "NC_landfill", "NC_other", 
    "NC_combustion_coal", "NC_combustion_msw"
]

df_combinations = pd.DataFrame(flattened_combinations, columns=columns)

# 添加 'num' 列
df_combinations.insert(0, 'num', range(0, len(df_combinations) ))

df_combinations.to_csv(r"E:\Test\Inputs\costs_SA\Cost_Combination_Schemes_95CL.csv",index=False)
```
